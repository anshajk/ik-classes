```python
# How to use groupby and agg

df.groupby('species').agg(
    {
        'sepal_length':'mean',
        'sepal_width':'mean',
        'petal_length':'median',
        'petal_width':['median','mean','count','sum']
    }
)
```

```python
print(df[df['species'] == 'setosa']['sepal_length'].mean())
print(df[df['species'] == 'versicolor']['sepal_length'].mean())
print(df[df['species'] == 'virginica']['sepal_length'].mean())
```

```python
grouped_df = df.groupby('species').agg(
    {
        'sepal_length':'mean',
        'sepal_width':'mean',
        'petal_length':'median',
        'petal_width': 'mean'
    }
)
```
