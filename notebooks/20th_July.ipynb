```python
clustering_df = df[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]
clustering_df.head()
```

```python
from sklearn.cluster import KMeans

from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()

scaled_data = scaler.fit_transform(clustering_df)
```

```python
scaled_data
```

```python
algorithm = KMeans(n_clusters=3)
```

```python
cluster_values = algorithm.fit_predict(scaled_data)
```

```python
cluster_values
```

```python
clustering_df.head()
```

```python
clustering_df['cluster'] = cluster_values
clustering_df.head()
```

```python
sns.set_palette('pastel')
sns.scatterplot(x='petal_length',y='petal_width', hue='cluster',data=clustering_df)
```
